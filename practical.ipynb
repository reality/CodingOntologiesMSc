We can load some of these interesting tables into pandas, and take a look:

So, here we have a few columns:

* row_id: This identifies the row in this table (boring

# Part 2: Querying Biomedical Ontologies

So before we talked about the Disease Ontology (DO), and how this stratifies different diseases and so on. We can use a handy piece of software called Komenti to look at different diseases and so on.

To run a query on a biomedical ontology, we can use Komenti, which links to an ontology repository called AberOWL (which you can also browse on the web from https://aber-owl.net/)

In the following command, we use komenti to 

* -q: This argument identifies the query you're running. Note: if the label is more than one word, it needs to additionally have a single quote around it.
* --class-mode
* --query-type

You can get a full-ish list of the query arguments [here](https://github.com/reality/Komenti)


By changing the query type, we can also retrieve more specific kinds of disease:

```python
!komenti query -q "cardiomyopathy" -o DOID --class-mode --query-type equivalent
```

We can also retrieve, from the Human Phenotype ontology, different phenotypes:

```python
!komenti query -q "hypertension" -o HP --class-mode --query-type subeq
```

You can also retrieve classes directly from their IRIs. You just have to wrap the argument with angled brackets (< and >). Try going to [AberOWL](https://aber-owl.net/ontology/HP) and picking a phenotype you're interested in, then insert it into the query below:

```python
!komenti query -q "<[INSERT IRI HERE]>" -o HP --class-mode --query-type equivalent
```

# Part 3: Linking EHR Entities to Ontologies

Now, this can take a little bit of work to parse and match up with the . Especially if we were interested in all of them. Thankfully, we can take this handy DOID<->ICD9 mapping I created earlier!

```python
!rm doid_icd_map*
!wget http://lokero.xyz/doid_icd_mappings.txt

doid_icd_map = pd.read_csv('doid_icd_mappings.txt', sep='\t')
doid_icd_map['icd9_code'] = doid_icd_map['icd9_code'].map(lambda icd: icd.replace('.', ''))

doid_icd_map
```

Very nice, now we can do another simple left join on our diagnoses table to add our maps

```python
diag_with_doid = diag_with_titles.merge(doid_icd_map, on="icd9_code", how="left")
print(str(len(diag_with_doid[diag_with_doid['doid'].notnull()])) + " out of " + str(len(diag_with_doid))) 

diag_with_doid
```

Depressingly, there were only 334 with a direct match! Thankfully we know that these things are matched.

```python
from difflib import SequenceMatcher
match = list()
for index, row in diag_with_doid.iterrows():
    if pd.isnull(row['doid']) and row['doid'] not in match:
      match.append(row['icd9_code'])
     
ii = 0
for code in match:
    longest_match = False
    longest_doid = False
    for i, dent in doid_icd_map.iterrows():
        if len(dent['icd9_code']) < len(code):
            if code.find(dent['icd9_code']) == 0:
                if not longest_match or longest_match and len(longest_match) < len(dent['icd9_code']):
                    longest_match = dent['icd9_code']
                    longest_doid = dent['doid']
    if longest_match:
        print('ye')
        print(longest_doid)
        diag_with_doid.loc[diag_with_doid.icd9_code == code, 'doid'] = longest_doid
    
    ii+=1
    print(str(ii)+'/'+str(len(match)))
```

# Part 4: Semantic characterisation

So now we can start to think about how we can build phenotype profiles, and use the underlying semantic information the ontologies to start to discern between 

```python
admissions = dict()
for index, row in diag_with_doid.iterrows():
    if row['hadm_id'] not in admissions:
        admissions[row['hadm_id']] = {
            'codes': list(),
            'contains_hd': False
        }
        
    if row['doid'] in hd_classes.keys():
        print('meow')
        admissions[row['hadm_id']]['contains_hd'] = True
    #else:
    admissions[row['hadm_id']]['codes'].append(row['doid'])

a = 0
with open('klar_input.tsv', 'w') as f:
    for aid, v in admissions.items():
      a+= 1
      print(a)
      groups = list()
      if v['contains_hd']:
        groups.append('HD')
        print('meow2')
      else:
        groups.append('NoHD')
    
      if len(groups) != 0:
        print('meow3')
        line = "\t".join((str(aid), ";".join(v['codes']), ";".join(groups)))
        print(line)
        f.write(line + '\n')
        
```

Now it's time to 

```python
labevents = pd.read_csv("mimic/LABEVENTS.csv")
labdefs = pd.read_csv("mimic/D_LABITEMS.csv")[['itemid','label']]
```

```python
labevents = labevents.merge(labdefs, on="itemid", how="left")
labevents = labevents[labevents['hadm_id'].notnull()]

labevents
```

So we have now all of these nice blood measurements, but how do we apply them to anything?

```python
blood_ranges = pd.read_csv('blood_ranges.tsv', sep='\t')
blood_ranges
```

So what we want to do is find people who have test scores outside of this 

```python
admission_phenotypes = dict()
for i, row in labevents.iterrows():
  if len(diag_with_doid[diag_with_doid['hadm_id'] == row['hadm_id']]) != 0:
    if row['hadm_id'] not in admission_phenotypes:
      admission_phenotypes[row['hadm_id']] = list()
    for z, ra in blood_ranges.iterrows():
      if row['itemid'] == ra['itemid']:
        try:
          value = float(row['value'].replace('<','').replace('>',''))
        except:
          continue
        if value < float(ra['lower']) and pd.notnull(ra['decreased_phen']) and ra['decreased_phen'] not in admission_phenotypes[row['hadm_id']]:
          admission_phenotypes[row['hadm_id']].append(ra['decreased_phen'])
        elif value > float(ra['upper']) and pd.notnull(ra['increased_phen']) and ra['increased_phen'] not in admission_phenotypes[row['hadm_id']]:
          admission_phenotypes[row['hadm_id']].append(ra['increased_phen'])
        
admission_phenotypes
```

```python
a = 0
with open('klar_input.tsv', 'w') as f:
    for aid, v in admissions.items():
      a+= 1
      print(a)
      groups = list()
      if v['contains_hd']:
        groups.append('HD')
      else:
        groups.append('NoHD')
    
      if len(groups) != 0 and len(admission_phenotypes[aid]) > 0:
        line = "\t".join((str(aid), ";".join(admission_phenotypes[aid]), ";".join(groups)))
        print(line)
        f.write(line + '\n')
```
